```python
def add_tags(df):
    for rating in df['Rating']:
        if rating == 1:
            df['Tags'] = random.choice(rating1)
        elif rating ==2:
            df['Tags'] = random.choice(rating2)
        elif rating ==3:
            df['Tags'] = random.choice(rating3)
        elif rating ==4:
            df['Tags'] = random.choice(rating4)
        elif rating ==5:
            df['Tags'] = random.choice(rating5)
    return df
```

```python
rating_df = add_tags(rating_df)
```
